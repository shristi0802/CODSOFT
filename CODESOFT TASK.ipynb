```python
#AIRTHMETIC CALCULATOR
from tkinter import *
window = Tk()
window.title("Arithmetic calculator")
window.geometry('950x300')

def calculate():
    try:
        num1 = float(entry1.get())
        num2 = float(entry2.get())
    except ValueError:
        res_label.config(text="Please enter valid numbers")
        return
    
    operation = operation_var.get()

    if operation == "+":
        result = num1 + num2
    elif operation == "-":
        result = num1-num2
    elif operation == "*":
        result = num1*num2
    elif operation == "/":
        if num2 == 0:
            res_label.config(text="cannot divide by zero")
            return
        result = num1/num2
    else:
        res_label.config(text="please select an operation")
        return
    res_label.config(text=f"Result: {result}")

lb1 = Label(window, text="number1:")
lb1.grid(row=0, column=0, padx=10, pady=5)

entry1 = Entry(window)
entry1.grid(row=0, column=1, padx=10, pady=5)

lb2 = Label(window, text="number2:")
lb2.grid(row=1, column=0, padx=10, pady=5)

entry2 = Entry(window)
entry2.grid(row=1, column=1, padx=10, pady=5)

operation_var = StringVar()
operation_var.set("+")

lb3 = Label(window, text="Operation:")
lb3.grid(row=2, column=0, padx= 10, pady=5)

operations_frame = Frame(window)
operations_frame.grid(row=2, column=1, padx=10, pady=5)

radio_add = Radiobutton(operations_frame, text="+", variable = operation_var, value= "+")
radio_add.pack(side=LEFT)

radio_sub = Radiobutton(operations_frame, text="-", variable = operation_var, value= "-")
radio_sub.pack(side=LEFT)

radio_product = Radiobutton(operations_frame, text="*", variable = operation_var, value= "*")
radio_product.pack(side=LEFT)

radio_div = Radiobutton(operations_frame, text="/", variable = operation_var, value= "/")
radio_div.pack(side=LEFT)

submit_button = Button(window, text="Calculate", command=calculate)
submit_button.grid(row=3, column=0, columnspan=2, pady=10)

res_label = Label(window, text="Result: ")
res_label.grid(row=4, column=0, columnspan=2, pady=5)

window.mainloop()





```

```python
#GENERATTING A PASSWORD
import random
import string

# Prompt the user to specify the desired length of the password
length = int(input("Enter the desired length of the password: "))

def generate_password(length):
    # Define the characters to use in the password
    characters = string.ascii_letters + string.digits + string.punctuation
    
    # Create a list to hold the password characters
    password = []
    
    # Generate the password by randomly selecting characters
    for i in range(length):
        password.append(random.choice(characters))
    
    # Join the list into a string and return it
    return ''.join(password)

# Generate the password
password = generate_password(length)

# Display the generated password
print("Generated Password: ", password)
```

```python
#TO-DO LIST
def add_task(tasks, title, description, deadline=None):
    task_id = len(tasks) + 1
    task = {
        'id': task_id,
        'title': title,
        'description': description,
        'status': 'incomplete',
        'deadline': deadline
    }
    tasks.append(task)
    return task

def view_tasks(tasks):
    if not tasks:
        print("No tasks available.")
    for task in tasks:
        print(f"ID: {task['id']} - Title: {task['title']} - Status: {task['status']} - Deadline: {task['deadline']}")

def update_task(tasks, task_id, new_title=None, new_description=None, new_deadline=None):
    for task in tasks:
        if task['id'] == task_id:
            if new_title:
                task['title'] = new_title
            if new_description:
                task['description'] = new_description
            if new_deadline:
                task['deadline'] = new_deadline
            return task
    print("Task not found.")
    return None

def delete_task(tasks, task_id):
    for task in tasks:
        if task['id'] == task_id:
            tasks.remove(task)
            return task
    print("Task not found.")
    return None

def mark_task(tasks, task_id, status):
    for task in tasks:
        if task['id'] == task_id:
            task['status'] = status
            return task
    print("Task not found.")
    return None

def main():
    tasks = []
    while True:
        print("\nTo-Do List Application")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Update Task")
        print("4. Delete Task")
        print("5. Mark Task Complete")
        print("6. Exit")
        choice = input("Enter your choice: ")
        
        if choice == '1':
            title = input("Enter task title: ")
            description = input("Enter task description: ")
            deadline = input("Enter task deadline (optional): ")
            add_task(tasks, title, description, deadline)
        elif choice == '2':
            view_tasks(tasks)
        elif choice == '3':
            task_id = int(input("Enter task ID to update: "))
            new_title = input("Enter new title (optional): ")
            new_description = input("Enter new description (optional): ")
            new_deadline = input("Enter new deadline (optional): ")
            update_task(tasks, task_id, new_title, new_description, new_deadline)
        elif choice == '4':
            task_id = int(input("Enter task ID to delete: "))
            delete_task(tasks, task_id)
        elif choice == '5':
            task_id = int(input("Enter task ID to mark complete/incomplete: "))
            status = input("Enter status (complete/incomplete): ")
            mark_task(tasks, task_id, status)
        elif choice == '6':
            break
        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```
